what is ment by visualization?

Data visualization is the discipline of trying to understand data by placing it in a visual context so that patterns, trends and correlations that might not otherwise be detected can be exposed. Python offers multiple great graphing libraries that come packed with lots of different features. No matter if you want to create interactive, live or highly customized plots python has an excellent library for you. Matplotlib: low level, provides lots of freedom

->Pandas Visualization: easy to use interface, built on Matplotlib

Defining Exploratory Data Analys

Exploratory Data Analysis – EDA – plays a critical role in understanding the what, why, and how of the problem statement. It’s first in the order of operations that a data analyst will perform when handed a new data source and problem statement

univariate(one dimension)

In univariate we pick up the one feature and bassed on the feature we will detremine what the output is, for example we take the feature as product or mrp or kms and plot these on x- axis by using the univariate we are not easy to classify the data, some of the points may be overlapping beacuse of that just by using one feature we are not able clasify the outputs currectly so due to that we go with the bivariate and multivariate analysis.

Bivariate:(two dimension)

In bivariate analysis we will take atleat two features like mrp , kms or poduct , year so on. also in bivariate the points are overlapped and it will be soled by using ML algorithms , multivariate

## Count plot

```python
plt.figure(figsize=(7,4))
sns.countplot(data=camera,x="Brand")
plt.xticks(rotation=90)
plt.title('Brand Vs Frequency')
plt.show()
```

-> Most number of cameras are from SJCAM Brand in our data.   \
-> Least number of cameras are from Avyana, KOOLCRAFTS,1CAMPRO, Garundropsy and Elevea Brand in our data.

## Histogram Plot

```python
plt.hist(camera['Pixels'],rwidth=0.9)
plt.title('Pixels Vs Frequency')
plt.xlabel('Pixels')
plt.ylabel('Frequency')
```

## Histogram Plot

```python
plt.hist(camera['Resolution(MP)'],rwidth=0.63,color='m')
plt.title('Resolution(MP) Vs Frequency')
plt.xlabel('Resolution(MP)')
plt.ylabel('Frequency')
```

```python
max_min_pixels=pd.pivot_table(data=camera,index='Brand',values=['Pixels'],aggfunc=['min','max'])
```

```python
max_min_pixels
```

## Bar Plot

```python
plt.figure(figsize=(400,10))
max_min_pixels.plot(kind='bar')
plt.title('Brand Vs Pixels')
plt.xlabel('Brands')
plt.ylabel('Max and Min Pixels')
plt.show()
```

-> Cameras in our data which have highest Image Quality are Insta360,dji, Gopro, SJCAM, Ricoh.  \
-> Cameras in our data which have low Image Quality are Pelupa, Cospex.

## Scatter Plot

```python
plt.figure(figsize=(20, 5))
plt.scatter(camera['Brand'], camera['Price'])
plt.xticks(rotation=90,size=15)
plt.yticks(size=15)
plt.title('Brand Vs Price',size=30)
plt.xlabel('Brand',size=20)
plt.ylabel('Price',size=20)
plt.show()
```

-> Digitek branded camera has highest price in our data.  \
-> SrO branded camera has lowest price in our data. 

## Point Plot

```python
plt.figure(figsize=(10,3))
agg_df = camera.groupby('Brand')['Resolution(MP)'].agg(['min', 'max']).reset_index()
sns.pointplot(x='Brand', y='min', data=agg_df, color='red', label='Min Value')
sns.pointplot(x='Brand', y='max', data=agg_df, color='blue', label='Max Value')
plt.xticks(rotation=90)
plt.title("Brand Vs Resolution")
plt.ylabel("Max and Min Resolution")
plt.grid(True)
plt.legend()
plt.show()
```

-> Insta360 branded camera has highest resolution in our data.  \
-> 17 brands have low resolution in our data.
