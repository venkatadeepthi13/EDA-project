# Import necessary libraries

```python
import requests 
import re
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

# Requesting page and Web scraping

```python
price=[]
total_url = []
brand=[]
color=[]
pixels=[]
shooting_modes=[]
fps=[]
video_quality=[]
video_resolution=[]
shutter_type=[]
sensor_size=[]
camera_size_whdwe=[]
j=0
for i in range(1,8):
    url='https://www.flipkart.com/cameras/sports-action/pr?sid=jek,p31,s3q&q=cameras&otracker=categorytree&page={}'.format(i)
    page = requests.get(url)
    print(page)
    soup = BeautifulSoup(page.text)
    for i in soup.find_all("a",class_="_1fQZEK"):
        domain = re.findall("href=(\"/\w.+)\srel",str(i))[0][1:-1]
        url1 = "https://www.flipkart.com"+domain
        total_url.append(url1)
    for i in total_url:
        page = requests.get(i)
        print(page)
        soup = BeautifulSoup(page.text)
        p=soup.find_all("div",class_="_30jeq3 _16Jk6d")
        for i in p:
            print(i.text)
            if i.text:
                price.append(i.text)
            else:
                price.append(np.nan)
        c=0
        for i in soup.find_all("div",class_="_3k-BhJ"):
            c=c+1
            if c==2:
                br=re.findall("Brand(\w+)M",i.text)
                if br:
                    brand.append(str(br[0]))
                else:
                    brand.append(np.nan)
                co=re.findall("Brand\sColor\w+Color(\w+)E",i.text)
                if co:
                    color.append(str(co[0]))
                else:
                    color.append(np.nan)
                pi=re.findall("Effective\sPixels(\d+)\sMP",i.text)
                if pi:
                    pixels.append(str(pi[0]))
                else:
                    pixels.append(np.nan)
                sm=re.findall("Shooting\sModes(\w.+)",i.text)
                if sm:
                    shooting_modes.append(str(sm[0]))
                else:
                    shooting_modes.append(np.nan)
            elif c==3:
                frames_per_second=re.findall("Frames\sPer\sSecond(\w.+)V",i.text)
                if frames_per_second:
                    fps.append(str(frames_per_second[0]))
                else:
                    fps.append(np.nan)
                vq=re.findall("Video\sQuality(\w.+)Fr",i.text)
                if vq:
                    video_quality.append(str(vq[0]))
                else:
                    video_quality.append(np.nan)
                vr=re.findall("Video\sResolution(\w.+)Video\sQu",i.text)
                if vr:
                    video_resolution.append(str(vr[0]))
                else:
                    video_resolution.append(np.nan)
            elif c==4:
                st=re.findall("Shutter\sType(\w+)",i.text)
                if st:
                    shutter_type.append(str(st[0]))
                else:
                    shutter_type.append(np.nan)
            elif c==5:
                ss=re.findall("Sensor\sSize(\d+\/?\d+?\.?\d+?)",i.text)
                if ss:
                    sensor_size.append(str(ss[0]))
                else:
                    sensor_size.append(np.nan)
            elif c==9:
                d=re.findall("DimensionsWidth(\d.+)Height(\d.+)Depth(\d.+)Weight(\d.+)",i.text)
                print(d)
                if d:
                    camera_size_whdwe.append(str(d[0]))
                else:
                    camera_size_whdwe.append(np.nan)
        j=j+1
        print(j)
#print(brand,color,pixels,shooting_modes,frames_per_second,video_quality,video_resolution,shutter_type,sensor_size,camera_size_whdwe)
```

# Creating dataframe

```python
camera=pd.DataFrame({"Brand":brand,"Pixels":pixels,"Resolution":video_resolution,"Quality":video_quality,"Frames per second":fps,"Shutter":shutter_type,"Sensor size":sensor_size,"Color":color,"Camera size":camera_size_whdwe,"Shooting modes":shooting_modes,"Price":price})
```
